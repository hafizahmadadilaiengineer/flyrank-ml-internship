## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

### Research Question

Which content pages should be reviewed first because they show evidence of declining performance while still receiving meaningful search visibility?

### Decision

Prioritize which pages should enter the content review queue first.

### Who Acts?

SEO specialists and content editors.

### Action

The highest-ranked pages will be reviewed for possible content updates, metadata improvements, expansion, or monitoring.

### Cost of a Wrong Recommendation

If a low-value page is ranked too high, editors waste valuable time reviewing pages with little opportunity.

If a high-value page is ranked too low, important content opportunities may be missed, potentially resulting in continued traffic decline.

### Why Machine Learning?

Machine learning can combine multiple observable search and engagement signals simultaneously, allowing it to prioritize pages more effectively than simple rule-based approaches.


## 3. Quick look at the data (2-3 real numbers)

```python
from pathlib import Path
import pandas as pd

repo_root = Path("flyrank-ml-internship")
df = pd.read_csv(repo_root / "data/raw/content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)
print("\nTrend Direction:")
print(df["trend_direction"].value_counts())

print("\nAverage CTR:", df["ctr"].mean())
print("Average Impressions:", df["impressions_90d"].mean())
print("Average Sessions:", df["sessions_90d"].mean())
```

The starter dataset contains **30,000 content pages** with **44 features**, providing sufficient data for exploratory analysis and model development.

Among these pages, **16,262 are labeled as "down"**, indicating that declining content represents a significant portion of the dataset and provides enough examples for studying content refresh opportunities.

The dataset also has an **average of 5,200 impressions over the last 90 days** and an **average CTR of approximately 0.51%**, showing that many pages receive measurable search visibility while still leaving room for improvement. These observations suggest that ranking pages for content review is a meaningful and practical machine learning problem.
